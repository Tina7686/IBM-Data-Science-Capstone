```python
import random
from math import sin, pi

def estimate_pi(N):
    # Initialize variables
    l = 1
    d = 2
    K = 0  # Counter for the number of intersections

    # Perform N experiments
    for _ in range(N):
        # Step 2: Randomly sample y-coordinate
        y = random.uniform(0, 2)

        # Step 3: Randomly sample angle θ
        theta = random.uniform(0, 2 * pi)

        # Calculate the distance from the center of the needle to the nearest boundary
        distance_to_boundary = min(y, 2 - y)

        # Check if the needle intersects the boundary
        if distance_to_boundary < (l / 2) * abs(sin(theta)):
            K += 1

    # Step 6: Calculate the estimated value of π
    estimated_pi = (2 * l) / (2 * (K / N))
    return estimated_pi

if __name__ == "__main__":
    N = 10000  # Number of experiments
    estimated_pi = estimate_pi(N)
    print(f"Estimated value of π is: {estimated_pi}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the dimensionality (n)
dimensions = [2, 3, 1000]

# Number of samples
num_samples = 10000

# Initialize the random seed for reproducibility
np.random.seed(0)

# Function to sample random vectors and compute their ℓ2 norm
def sample_norm(dim, num_samples):
    samples = np.random.normal(0, 1, size=(num_samples, dim))
    norms = np.linalg.norm(samples, axis=1)
    return norms

# Create subplots for each dimension
fig, axes = plt.subplots(1, len(dimensions), figsize=(15, 5))

for i, dim in enumerate(dimensions):
    norms = sample_norm(dim, num_samples)

    # Plot the histogram
    axes[i].hist(norms, bins=50, density=True, alpha=0.6, color='blue')
    axes[i].set_title(f'n = {dim}')
    axes[i].set_xlabel('Norm')
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)  # Set the random seed for reproducibility

# Define the dimensions
dimensions = [2, 3, 1000]

# Generate random vectors and calculate their lengths
lengths = []
for n in dimensions:
    g = np.random.normal(0, 1, size=(n,))
    length = np.linalg.norm(g)
    lengths.append(length)

# Plot the histograms
plt.figure(figsize=(8, 6))
for i, n in enumerate(dimensions):
    plt.subplot(2, 2, i+1)
    plt.hist(lengths[i], bins=50, density=True)
    plt.title(f'n = {n}')
    plt.xlabel('Length')
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_vector(n):
    return np.random.normal(0, 1, n)

def plot_histogram(n):
    num_samples = 10000
    random_vectors = np.zeros(num_samples)
    
    for i in range(num_samples):
        random_vector = generate_random_vector(n)
        random_vectors[i] = np.linalg.norm(random_vector)
    
    plt.hist(random_vectors, bins=50, density=True)
    plt.xlabel('Length')
    plt.ylabel('Probability Density')
    plt.title(f'Histogram of Random Vector Lengths (n = {n})')
    plt.show()

# Visualize histogram for n = 2
plot_histogram(2)

# Visualize histogram for n = 3
plot_histogram(3)

# Visualize histogram for n = 1000
plot_histogram(1000)
```
